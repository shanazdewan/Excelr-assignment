### Normalization of the data

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

## KMeans Clustering

```python
scaler = StandardScaler()
```

```python
scaled_df = scaler.fit_transform(crime.iloc[:,1:])
scaled_df
```

```python
fig = plt.figure(figsize=(10,8))
WCSS = []
for i in range(1,10):
    km = KMeans(n_clusters= i)
    km.fit(df_norm)
    WCSS.append(km.inertia_)
plt.plot(range(1,10),WCSS)
plt.title("Elbow Method")
plt.ylabel("WCSS")
plt.xlabel("Number of clusters")
plt.show()
```

```python
k_cluster = KMeans(n_clusters=4)
y_k = k_cluster.fit_predict(df_norm)
y_k
```

```python
c_k = pd.Series(y_k)
c_k
```

```python
crime_Kmeans = crime.copy()
crime_Kmeans["K_clust"] = c_k
crime_Kmeans
```

```python
sklearn.metrics.silhouette_score(scaled_df,y_k)
```

```python
crime.iloc[:,1:].groupby(crime_Kmeans.K_clust).mean()
```

## DBSCAN

```python
dbscan = DBSCAN(eps = 1, min_samples = 4)
dbscan.fit(scaled_df)
```

```python
dbscan.labels_
```

```python
c_d = pd.Series(dbscan.labels_)
c_d
```

```python
crime_dbscan = crime.copy()
crime_dbscan["D_clust"] = c_d
crime_dbscan
```

```python
sklearn.metrics.silhouette_score(scaled_df,c_d)
```

```python
crime.iloc[:,1:].groupby(crime_dbscan.D_clust).mean()
```

## Hierarchical Clustering

```python
z = linkage(df_norm,method="average",metric="euclidean")
plt.figure(figsize=(15,5))
plt.title("Hierarchical Clustering Denfrogram")
sch.dendrogram(z)
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters=5,linkage="complete",affinity="euclidean").fit(df_norm)
```

```python
hc.labels_
```

```python
cluster_hc = pd.Series(hc.labels_)
cluster_hc
```

```python
sklearn.metrics.silhouette_score(scaled_df,cluster_hc)
```

```python
crime_HC = crime.copy()
```

```python
crime_HC["H_clust"] = cluster_hc
```

```python
crime_HC
```

```python
crime.iloc[:,1:].groupby(crime_HC.H_clust).mean()
```
