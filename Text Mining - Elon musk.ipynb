```python
def clean_text(text):
    
    text=re.sub('@[A-Za-z0-9]+','',str(text))
    text=re.sub('#','',str(text)) 
    text=re.sub('RT[\s]+','',str(text))
    text=re.sub('\n\n','',str(text)) 
    text=text.lower()
    text=re.sub('https?:\/\/\S+','',str(text)) 
    text=text.translate(str.maketrans('', '', string.punctuation))

    return text

data['Text']=data["Text"].apply(clean_text)
data
```

```python
elon=data.Text
elon_string=" ".join(elon)
elon_string
```

```python
text_tokens=word_tokenize(elon_string)
print(text_tokens)
```

```python
with open ("C:\\Users\\Shanaz Dewan\\Downloads\\stop.txt","r") as sw:
    stop_words=sw.read()
stop_words=stop_words.split('\n') 
```

```python
tweets_no_sw=[word for word in text_tokens if word not in stop_words]
tweets_no_sw
```

```python
tf=TfidfVectorizer()
text_tf=tf.fit_transform(tweets_no_sw)
```

```python
feature_names=tf.get_feature_names()
dense=text_tf.todense()
denselist=dense.tolist()
df=pd.DataFrame(denselist,columns=feature_names)
df
```

```python
wordslist=" ".join(df)
```

```python
from textblob import TextBlob
def sentiment_analysis(ds):
    sentiment = TextBlob(ds["Text"]).sentiment
    return pd.Series([sentiment.subjectivity, sentiment.polarity])

data[["subjectivity", "polarity"]] = data.apply(sentiment_analysis, axis=1)
data
```

```python
def analysis(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
    
data["analysis"] = data["polarity"].apply(analysis)
data
```
